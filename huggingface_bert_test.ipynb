```python
from huggingface_hub import model_info

# BERT base uncased 모델 정보 불러오기
info = model_info("bert-base-uncased") # 대소문자 구별 안함

# 전체 메타데이터 보기
print(info.cardData)

# 특정 항목 추출
if "model_size" in info.cardData:
    print("모델 파라미터 수:", info.cardData["model_size"])
if "datasets" in info.cardData:
    print("학습 데이터셋:", info.cardData["datasets"])

print(f"\ninference:, {info.inference}")
print(f"\nconfig:, {info.config}")
print(f"\nlibrary_name:, {info.library_name}")
print(info.safetensors)
```

# LLM을 이용하여 텍스트 생성하기

```python
# Phi-3 모델과 호환성 때문에 transformers 4.48.3 버전을 사용합니다.
!pip install transformers==4.48.3
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# 모델과 토크나이저 로드하기
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda", # GPU를 사용하겠다. (cuda 모델환경에서 돌아감)
    torch_dtype="auto",
    trust_remote_code=False, # 캐싱문제 발생: True -> False로 바꿈
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
pip install -U "transformers>=4.44" "accelerate>=0.34" "torch>=2.2"
```

```python
from transformers import pipeline

# 파이프라인 생성
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False, # True일 때랑 False일 때랑 비교해보기
    max_new_tokens=500,
    do_sample=False
)

# 프롬프트(사용자 입력/쿼리)
messages = [
    {"role": "user", "content": "Create a funny jike about chickens."}
]

# 출력 생성
output = generator(messages)
print(output[0]["generated_text"])
```
